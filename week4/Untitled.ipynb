```python
import pandas as pd
from datetime import datetime

df = pd.read_excel("hotelBookings.xlsx")
df
```

```python
def clean_1(df):
    if len(df[df.hotel != 'Resort Hotel']) == 0:
        print('column hotel is clean')
        return df
clean_1(df)
```

```python
def clean_2(df):
    if len(df[~df['is_canceled'].isin([0, 1])]) == 0:
        print('column hotel is clean')
        return df
clean_2(df)
```

```python
def clean_3(df):
    if len(df[df.lead_time < 0]) == 0 and len(df[df.lead_time%1 != 0]) == 0:
        print("column lead_time is clean")
        return df
clean_3(df)
```

```python
def clean_4(df):
    #vervang alle jaren hoger dan het huidig jaar door 2015, want dat is het enige jaar dat voorkomt
    nono = df[df.arrival_date_year > datetime.now().year].index
    df.loc[df.index.isin(nono),'arrival_date_year'] = 2015
    if len(df[df.arrival_date_year > datetime.now().year]) == 0:
        print("column arrival_date_year is clean")
        return df
clean_4(df)
```

```python
def clean_5(df):
    #vervang alle rijen die niet
    nono = df[~df.arrival_date_month.isin(['Januari','Februari','March','April','May','June','July','August','September','October','November','December'])].index
    df.loc[df.index.isin(nono),'arrival_date_month'] = 'July'
    if len(df[df.arrival_date_year > datetime.now().year]) == 0:
        print("column arrival_date_month is clean")
        return df
clean_5(df)
```

```python
def clear_6(df):
    if len(df[~df.arrival_date_week_number.isin(range(0,51))]) == 0:
        print("column arrival_date_week_number is clean")
        return df
```

```python
def clear_7(df):
    if len(df[~df.arrival_date_day_of_month.isin(range(1,32))]) == 0:
        print("column arrival_date_day_of_month is clean")
        return df
clear_7(df)
```

```python
import multiprocessing

def calculate_optimal_range(start, end):
    optimal = 1
    for i in range(start, end):
        if optimal < i * 7 - i + 3:
            optimal = i * 7 - i + 3
    return optimal,i

if __name__ == "__main__":
    num_cores = multiprocessing.cpu_count()
    start = 0
    end = 100

    # Split the range into chunks for parallel processing
    chunk_size = (end - start) // num_cores
    chunks = [(i, i + chunk_size) for i in range(start, end, chunk_size)]

    # Create a pool of worker processes
    pool = multiprocessing.Pool(processes=num_cores)

    # Calculate the optimal values in parallel
    results = pool.starmap(calculate_optimal_range, chunks)

    # Close the pool and wait for all processes to finish
    pool.close()
    pool.join()

    # Find the overall optimal value
    overall_optimal = max(results)

    print("Overall optimal value:", overall_optimal)
```
